## Hyperparameter Optimization

Returns the configuration with the best average return over multiple trials with the same settings

#### Load Data

```python
trials = load_pickle('../results/experiment_2/hpo/PARAMETRIC_MASKED.pkl')
trials.head()
```

#### Remove index number from experiment tag

```python
trials['experiment_tag'] = trials['experiment_tag'].apply(lambda x: ','.join(str(x).split(',')[1:]))
trials.head()
```

#### Sort according to mean of experiment tags

```python
results = trials[['episode_reward_mean', 'experiment_tag']].groupby('experiment_tag').mean().sort_values(by='episode_reward_mean', ascending=False).reset_index()
print('best results:')
print(results.head())
best_result = results['experiment_tag'].iloc[0]
```

#### Get hyperparameters of best average result

```python
best_setup = trials[trials['experiment_tag'].str.endswith(best_result)]
print(len(best_setup))
list(best_setup.iloc[0].index)
```
